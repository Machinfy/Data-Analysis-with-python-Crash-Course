#### Merge & Concat

```python
df1 = pd.DataFrame({"A":[1,2,5],'B':[8,5,2]})
df2 = pd.DataFrame({"A":[1,2,3],'c':[2,4,6]})
```

```python
df1.head()
```

```python
df2.head()
```

```python
pd.concat([df1,df2])
```

```python
df1.merge(df2, how='left',on='A')
```

```python
df1.merge(df2, how='right',on='A')
```

```python
df1.merge(df2, how='outer',on='A')
```

```python
df1.merge(df2, how='outer',on='A')
```

```python
df1.merge(df2, how='inner',on='A')
```
